# Natural Join

```python
%%sql
select *
from product natural join vendor
```

```python
%%sql
select *
from product natural join vendor
order by v_code
```

```python
%%sql
select * 
from customer natural join vendor
```

# join using

```python
%%sql
select *
from product join vendor using(v_code)
order by v_code
```

```python
%%sql
select * 
from customer join vendor using (v_code)
```

# join on

```python
%%sql
select *
from product join vendor on product.v_code = vendor.v_code
order by v_code
```

```python
%%sql
select * 
from customer join vendor on customer.cus_areacode = vendor.v_areacode
```

# outer joins

```python
%%sql
select product.*, vendor.v_name
from product left join vendor on product.v_code = vendor.v_code
```

```python
%%sql
select product.*, vendor.v_name
from product right join vendor on product.v_code = vendor.v_code
```

```python
%%sql
select product.*, vendor.v_code, vendor.v_name
from product full outer join vendor on product.v_code = vendor.v_code
order by v_code
```

# aggregate processing

```python
%%sql
select sum(line_units * line_price) as inv_total
from line
where inv_number = 1006
```

```python
%%sql
select count(v_code) as `outside products`, count(p_code) as `total products`, p_code
from product
```

```python
%%sql
select sum(line_units * line_price) as inv_total,
       inv_number
from line
group by inv_number
```

```python
%%sql
select invoice.inv_number, 
       cus_fname || ' ' || cus_lname as `Customer Name`, 
       sum(line_units * line_price) as inv_total  
from line
join invoice on line.inv_number = invoice.inv_number
join customer on customer.cus_code = invoice.cus_code
group by invoice.inv_number
```

```python
%%sql
select invoice.inv_number, 
       cus_fname || ' ' || cus_lname as `Customer Name`, 
       sum(line_units * line_price) as inv_total  
from line
join invoice on line.inv_number = invoice.inv_number
join customer on customer.cus_code = invoice.cus_code
group by invoice.inv_number
having inv_total > 100
```

```python
%%sql
select invoice.inv_number, 
       cus_fname || ' ' || cus_lname as `Customer Name`, 
       sum(line_units * line_price) as inv_total  
from line
join invoice on line.inv_number = invoice.inv_number
join customer on customer.cus_code = invoice.cus_code
where line_units * line_price > 100
group by invoice.inv_number
```

# attribute list subquery

```python
%%sql
select avg(inv.inv_total)
from (select sum(line_units * line_price) as inv_total
      from line group by inv_number) as inv
```

```python
%%sql
select invoice.inv_number, 
       cus_fname || ' ' || cus_lname as `Customer Name`, 
       sum(line_units * line_price) as inv_total,
       (select avg(inv.inv_total)
       from (select sum(line_units * line_price) as inv_total
      from line group by inv_number) as inv) as avg_inv_total
from line
join invoice on line.inv_number = invoice.inv_number
join customer on customer.cus_code = invoice.cus_code
group by invoice.inv_number
```
