```python
%%sql
select *
from actor 
join cast as c on actor.id = c.actor_id
join movie on movie.id = c.movie_id
where movie.name like '%shrek%'
```

```python
%%sql
select count(*)
from actor 
join cast as c on actor.id = c.actor_id
join movie on movie.id = c.movie_id
where actor_id in (select distinct a.id 
                   from actor as a 
                    join cast as c2 on c2.actor_id = a.id
                    join movie as m on c2.movie_id = m.id
                   where m.name like 'shrek%')
```

```python
%%sql
select *
from movie
where movie.rank > all(select m.rank from movie as m where m.name like 'shrek%')
```

```python
%%sql
select *
from actor
where exists (select *  from actor as a
join cast as c on a.id = c.actor_id
join movie on movie.id = c.movie_id where a.id = actor.id and movie.name like 'shrek%')
```

```python
%%sql
select actor.* 
from actor 
join cast as cas on actor.id == cas.actor_id
join movie on cas.movie_id == movie.id
where movie.year > 1850 and movie.year < 1900
```

```python
%%sql
SELECT actor.*
FROM actor
JOIN cast as c ON actor.id = c.actor_id 
JOIN movie ON c.movie_id = movie.id 
WHERE movie.year >= 1850 and movie.year < 1900
INTERSECT
SELECT actor.*
FROM actor
JOIN cast as c ON actor.id = c.actor_id 
JOIN movie ON c.movie_id = movie.id 
WHERE movie.year >= 1900 and movie.year < 1950
```
