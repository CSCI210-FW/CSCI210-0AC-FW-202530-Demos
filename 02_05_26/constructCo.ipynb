# Select the first name, last name and hire date for all employees with the job code 504.

```python
%%sql
select emp_fname, emp_lname, emp_hiredate from employee where job_code = '504'
```

# Select first name and last name combined for all employees

```python
%%sql
select distinct emp_fname || ' ' ||  emp_lname as `Employee Name`
from employee
```

# Select the first name, last name and hire date for all employees with the job code 510 who have 16 years or less on the job.

```python
%%sql
select emp_fname, emp_lname, emp_hiredate
from employee
where (job_code == '510') and (emp_years <= '16')
```

```python
%%sql
select emp_fname, emp_lname, emp_hiredate from employee where job_code = '510' and emp_years < 16
```

```python
%%sql
select emp_fname, emp_lname, emp_hiredate
from employee 
where 
emp_hiredate >= date('now', '-16 years')
```
